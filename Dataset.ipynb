```python
import os
import torch
import torchvision.transforms as transforms
from PIL import Image
from matplotlib import pyplot as plt
```

```python
ROOT = "Dataset"
```

```python
for k in os.listdir(ROOT):
    print(k)
    counter = {
        "airs": 0,
        "inria": 0,
        "mbd": 0
        
    }
    heat_map = torch.zeros(1, 512, 512)
    for mask in os.listdir(os.path.join(ROOT, k, "Labels")):
        im = Image.open(os.path.join(ROOT, k, "Labels", mask)).convert("L")
        im = transforms.ToTensor()(im)
        heat_map += im

        if "airs" in mask:
            counter["airs"] += 1
        elif "inria" in mask:
            counter["inria"] += 1
        elif "mbd" in mask:
            counter["mbd"] += 1

    print(torch.max(heat_map), torch.min(heat_map))
    print(counter)
    
    plt.imshow(heat_map[0])
    plt.axis("off")
    plt.show()
```
